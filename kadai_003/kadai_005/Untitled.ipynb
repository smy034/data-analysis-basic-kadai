```python
import numpy as np
import matplotlib.pyplot as plt

# 描画領域を作成（2行3列）
fig, axes = plt.subplots(2, 3, figsize=(10, 6))

# -----------------------------
# 1. 折れ線グラフ
# -----------------------------
x1 = [0, 1, 2, 3]
y1 = [1, 2, -5, 2]

axes[0, 0].plot(x1, y1)
axes[0, 0].set_title("Line Graph")

# -----------------------------
# 2. sin関数
# -----------------------------
x2 = np.arange(0, 10, 0.1)
y2 = 2 * np.sin(2 * x2) + 2

axes[0, 1].plot(x2, y2)
axes[0, 1].set_title("Sin Function")

# -----------------------------
# 3. ヒストグラム
# -----------------------------
data = np.random.normal(15, 5, 2000)

axes[0, 2].hist(data, bins=10)
axes[0, 2].set_title("Histogram")

# -----------------------------
# 4. 散布図
# -----------------------------
x4 = np.random.normal(4, 2, 50)
y4 = np.random.normal(4, 2, 50)

axes[1, 0].scatter(x4, y4)
axes[1, 0].set_title("Scatter")

# レイアウト調整
plt.tight_layout()

# 表示
plt.show()
```
