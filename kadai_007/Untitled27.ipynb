```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

# CSVファイルの読み込み
df = pd.read_csv("sample_pandas_6.csv")

# -----------------------------
# ① カテゴリー列のユニークな出現頻度をカウント
# -----------------------------

# カテゴリー列の出現頻度を集計
category_counts = df['カテゴリー'].value_counts()

# 結果を表示
print("【カテゴリー別 出現頻度】")
print(category_counts)

# -----------------------------
# 棒グラフを表示
# -----------------------------

category_counts.plot(kind='bar')

# グラフタイトル
plt.title('カテゴリー別 出現頻度')

# X軸ラベル
plt.xlabel('カテゴリー')

# Y軸ラベル
plt.ylabel('件数')

# グラフ表示
plt.show()

# -----------------------------
# ② 商品番号ごとに価格を集計
# -----------------------------

# 商品番号ごとの価格合計
group_sum = df.groupby('商品番号')['価格'].sum()

print("【商品番号ごとの価格合計】")
print(group_sum)

# -----------------------------
# 基本統計量を表示
# -----------------------------

group_describe = df.groupby('商品番号')['価格'].describe()

print("【商品番号ごとの基本統計量】")
print(group_describe)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib

df = pd.read_csv("sample_pandas_6.csv")

# 列名確認
print(df.columns)

# ① 商品名ごとの出現頻度
item_counts = df['商品名'].value_counts()

print("【商品名ごとの出現頻度】")
print(item_counts)

item_counts.plot(kind='bar')
plt.title('商品名ごとの出現頻度')
plt.xlabel('商品名')
plt.ylabel('件数')
plt.show()

# ② 商品番号ごとの注文数合計
group_sum = df.groupby('商品番号')['注文数'].sum()

print("【商品番号ごとの注文数合計】")
print(group_sum)

# 商品番号ごとの注文数の基本統計量
group_describe = df.groupby('商品番号')['注文数'].describe()

print("【商品番号ごとの注文数の基本統計量】")
print(group_describe)
```
